```python
import pandas as pd;
import numpy as np;
import matplotlib.pyplot as plt;
```

```python
%matplotlib inline
```

```python
x= np.array([1,2,3,4,5]);
y= np.array([5,7,9,11,13]);
l_rate = 0.08
iteration = 1000;
b_s = m_s = 0;
n = len(x)
```

```python
plt.scatter(x,y,color="red",marker='+',linewidth='5')
```

```python
cus_pre = np.array([]);
for i in range(iteration):
    #y=mx+b
    #y_pre = y predicted 
    y_pre = m_s * x + b_s;
    #mse = mean square error
    mse_or_cost= (1/n)*sum((y - y_pre)**2);
    # partial derivative
    d_m = -(2/n) * sum(x*(y - y_pre));
    d_b = -(2/n) * sum(y-y_pre);
    # Slope finding
    m_s = m_s - l_rate * d_m;
    b_s = b_s - l_rate * d_b;
    np.append(cus_pre,y_pre);
    print('ms {} bs {} cost {}'.format(m_s,b_s,mse_or_cost))
    plt.plot()
print('cus_pre',cus_pre,len(cus_pre))    
```
